```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("music_data_augmented.csv")
df.head()
```

```python
tracks = pd.read_csv("tracks.csv")
tracks.head()
```

```python
tracks['artiststring'] = tracks['artists'].apply(lambda x: x.replace("['", "").replace("']", "").replace("', '",", ")).apply(str.lower)
tracks["name"] = tracks['name'].map(lambda x: x.lower() if isinstance(x,str) else x).astype(str)
```

```python
track_cols = ["artiststring", "name", "popularity"]
sub_tracks = tracks[track_cols]
sub_tracks
```

```python
final_df = df.merge(sub_tracks, left_on=['artist_name', 'track_name'], right_on=['artiststring', 'name'], how='left')
final_df['popularity'] = final_df['popularity'].fillna(0)
final_df
```

```python
final_df.to_csv('music_data_with_popularity.csv')
```
